# BAR PLOT

```python
plt.bar([0,1,2],[10,20,15])
```

```python
x_cor = np.array([0,1,2])
# x_cor_another = [1,2,3]
plt.bar(x_cor,[20,10,45], width = 0.2, color = 'blue', label = 'H', tick_label = ["x","y","z"])
plt.bar(x_cor + 0.5,[35,3,15], width = 0.2, color = 'red', label = 'G', tick_label = ["a","b","c"])
plt.xlim(-0.5,3)
plt.ylim(0,50)
plt.show()
```

# PIE CHART

```python
plt.style.use("seaborn")
sub = "maths", "eng", "phy", "hist"
w = [20, 10, 15, 5]
plt.pie(w, labels = sub)
plt.show()
```

```python
plt.style.use("seaborn")
sub = "maths", "eng", "phy", "hist"
w = [20, 10, 15, 5]
plt.pie(w, labels = sub, explode = (0,0,0.5,0))
plt.show()
```

```python
plt.style.use("seaborn")
sub = "maths", "eng", "phy", "hist"
w = [20, 10, 15, 5]
plt.pie(w, labels = sub, explode = (0,0,0.5,0), shadow = True, startangle = 90)
plt.show()
```

```python
plt.style.use("seaborn")
sub = "maths", "eng", "phy", "hist"
w = [20, 10, 15, 5]
plt.pie(w, labels = sub, explode = (0.2,0,0.5,0), shadow = True, startangle = 90, autopct="%1.1f%%")
plt.show()
```

# HISTOGRAM

```python
xsn = np.random.randn(100)
sigma = 5
u = 20
x = xsn*sigma + u
x = np.round(x)
plt.hist(x)
```

```python
xsn = np.random.randn(1000)
sigma = 5
u = 20
x = xsn*sigma + u
x = np.round(x)
plt.hist(x,bins=100)
```

```python
xsn = np.random.randn(1000)
sigma = 5
u = 20
x = xsn*sigma + u
x = np.round(x)
plt.hist(x,bins=100)
x1 = xsn*6 + 40
plt.hist(x1, alpha = 0.5) #Alpha represents the level of opaqueness in the graph
plt.show()
```
