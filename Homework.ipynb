```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mnist_data = pd.read_csv("D:\\Data Science Classwork\\mnist_test.csv")
mnist_data.head()
```

```python
np.random.seed(10)
data = mnist_data.values
np.random.shuffle(data)
```

```python
x = data[:,1:]
y = data[:,0]
```

```python
def draw(x,y,i):
    plt.imshow(x[i].reshape(28,28), cmap = "gray")
    plt.title("Label" + str(y[i]))
```

```python
draw(x,y,1)
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x[i].reshape(28,28),cmap="gray")
    plt.axis("off")
```
